```python
# BURADA VERİSETİMİZİ %70 EĞİTİM %30 OLACAK ŞEKİLDE RASTGELE AYIRIYORUZ

import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

VeriSeti = pd.read_csv('veriSeti.csv')

VeriSeti.columns = ['pregnant', 'glucose', 'blood pressure', 'skin thickness', 'insulin', 'bmi', 'diabetes pedigree', 'age', 'class']

X = VeriSeti.drop('class', axis=1)
y = VeriSeti['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# Çoklu Doğrusal Regresyon Analizi

from sklearn.linear_model import LinearRegression

# Modeli oluşturma
linear_reg = LinearRegression()

# Modeli eğitme
linear_reg.fit(X_train, y_train)

# Katsayıları elde etme
coefficients = linear_reg.coef_

# Katsayıları raporlama
for i, col in enumerate(X.columns):
    print(f'{col}: {coefficients[i]}')

# Test kümesi için tahmin yapma
y_pred_linear = linear_reg.predict(X_test)

# Performans metriklerini hesaplama
linear_accuracy = accuracy_score(y_test, y_pred_linear.round())
linear_conf_matrix = confusion_matrix(y_test, y_pred_linear.round())
linear_classification_report = classification_report(y_test, y_pred_linear.round())

print("Accuracy Score (Çoklu Doğrusal Regresyon):", linear_accuracy)
print("Confusion Matrix (Çoklu Doğrusal Regresyon):\n", linear_conf_matrix)
print("Classification Report (Çoklu Doğrusal Regresyon):\n", linear_classification_report)
```

```python
# Multinominal Lojistik Regresyon Analizi

from sklearn.linear_model import LogisticRegression

# Modeli oluşturma
logistic_reg = LogisticRegression(multi_class='multinomial', solver='newton-cg')

# Modeli eğitme
logistic_reg.fit(X_train, y_train)

# Katsayıları elde etme
coefficients = logistic_reg.coef_

# Katsayıları raporlama
for i, col in enumerate(X.columns):
    print(f'{col}: {coefficients[0][i]}')

# Test kümesi için tahmin yapma
y_pred_logistic = logistic_reg.predict(X_test)

# Performans metriklerini hesaplama
logistic_accuracy = accuracy_score(y_test, y_pred_logistic)
logistic_conf_matrix = confusion_matrix(y_test, y_pred_logistic)
logistic_classification_report = classification_report(y_test, y_pred_logistic)

print("Accuracy Score (Multinominal Lojistik Regresyon):", logistic_accuracy)
print("Confusion Matrix (Multinominal Lojistik Regresyon):\n", logistic_conf_matrix)
print("Classification Report (Multinominal Lojistik Regresyon):\n", logistic_classification_report)
```

```python
print("Accuracy Score (Çoklu Doğrusal Regresyon):", linear_accuracy)
print("Accuracy Score (Multinominal Lojistik Regresyon):", logistic_accuracy)
```

Multinominal lojistik regresyon modeli ile elde edilen denklemde ise her bir sınıf için ayrı bir denklem bulunur. Ancak, performans metriklerini hesaplarken sadece doğruluk (accuracy) kullanılabilir çünkü çoklu sınıfların tahmin edilmesi söz konusu. Diğer metrikler (örneğin, hassasiyet, duyarlılık, f1-score) her bir sınıf için ayrı ayrı değerlendirilmelidir.
