```python
import numpy as np
import pandas as pd

veriler=pd.read_csv('veriSeti.csv')

# Minimum ve maksimum değerleri bulun
min_val = veriler.min()
max_val = veriler.max()

# Min-Max Normalizasyonu
normalized_data = (veriler - min_val) / (max_val - min_val)


print("\nNormalleştirilmiş Veriler:")
print(normalized_data)
```

```python
import numpy as np
import pandas as pd

# CSV dosyasını oku
veriler = pd.read_csv('veriSeti.csv')

# Z-Score Normalizasyonu
normalized_data = (veriler - veriler.mean()) / veriler.std()

print("\nZ-Score Normalizasyonu (Standardizasyon) ile Normalleştirilmiş Veriler:")
print(normalized_data)
```

```python
import numpy as np
import pandas as pd

# CSV dosyasını oku
veriler = pd.read_csv('veriSeti.csv')

# Medyan Normalizasyonu
normalized_data = (veriler - veriler.median()) / (veriler.quantile(0.75) - veriler.quantile(0.25))

print("\nMedyan Normalizasyonu ile Normalleştirilmiş Veriler:")
print(normalized_data)
```

```python
import numpy as np
import pandas as pd

def sigmoid_normalizasyon(x):
    return 1 / (1 + np.exp(-x))

# CSV dosyasını oku
veriler = pd.read_csv('veriSeti.csv')

# Sigmoid Normalizasyonu
normalized_data = sigmoid_normalizasyon(veriler)

print("\nSigmoid Normalizasyonu ile Normalleştirilmiş Veriler:")
print(normalized_data)
```
